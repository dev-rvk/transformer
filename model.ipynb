```python
import torch
from tokenizer.tokenizer import Tokenizer
from embedding.embedding import InputEmbedding, AbsolutePositionalEncoder, PositionalEncoding
```

```python
VOCAB_SIZE = 100256
D_MODEL = 512
```

```python
# take input
input_text = "!"
```

```python
tokenizer = Tokenizer()
input_encoder = InputEmbedding(vocab_size=VOCAB_SIZE, d_model=D_MODEL)
positional_encoder = PositionalEncoding(d_model=D_MODEL, max_len=512)
```

```python
tokens = torch.tensor(tokenizer.encode(input_text))
tokens
```

```python
input_embedding = input_encoder(tokens)
input_embedding
```

```python
embeddings = positional_encoder(input_embedding)
embeddings
```
